```python
import pandas as pd
import numpy as np
import cv2
```

```python
big5 = pd.read_excel('../data/BFD/raw_big5.ods')
big5.index += 1
```

```python
print(big5.shape)
big5.sample(3)
```

```python
trait_map = {
    'open':'Openness', 
    'cons':'Conscientiousness',
    'extr':'Extraversion',
    'agre':'Agreeableness',
    'neur':'Neuroticism'
}

trait_type_map = {
    'minus':'reduced', 
    'norm':'Original',
    'plus':'enhanced'
}
traits_type = ['minus','norm','plus']
traits = ['open','cons','extr','agre','neur']
```

```python
data = {}
for i in range(1,41):
    for trait_type in traits_type:
        traits_list = []
        for trait in traits:
                traits_list.append(np.float32(big5.loc[i][trait+"_norm"]))
        if trait_type == 'norm':        
            filename = str(i) + '_' + trait_type_map[trait_type] + '.png'
            data[filename] = traits_list
        else:
            for j, trait in enumerate(traits):
                tmp_traits_list = traits_list.copy()
                filename = str(i) + '_' + trait_map[trait] + '_' + trait_type_map[trait_type] + '.png'
                tmp_traits_list[j] = np.float32(big5.loc[i][trait+"_"+trait_type])
                data[filename] = tmp_traits_list
```

```python
final_df = pd.DataFrame.from_dict(data, orient='index', columns=['ValueOpenness','ValueConscientiousness','ValueExtraversion','ValueAgreeableness','ValueNeurotisicm'])
final_df.index.name = 'ImageName'
final_df = ((final_df-1)/4)
final_df
```

```python
final_df.to_csv('bigfive_labels.csv')
```
